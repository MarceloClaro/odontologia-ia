```python
# ============================================================
# 4. Geração e visualização de radiografia sintética
# ============================================================
import matplotlib.pyplot as plt
import numpy as np

# Gera uma radiografia panorâmica sintética
rng = np.random.default_rng(42)
img = rng.normal(40, 20, (512, 1024)).clip(0, 255).astype(np.uint8)

# Adiciona estruturas simulando dentes
import cv2
for i in range(14):
    center_x = int(100 + i * 65 + rng.integers(-15, 15))
    center_y = int(200 + 80 * np.sin(i * 0.5))
    cv2.ellipse(img, (center_x, center_y), (20, 12),
                rng.uniform(-20, 20), 0, 360, rng.integers(120, 200), -1)

# Aplica CLAHE
from odontoia.data.preprocessing import clahe_equalization
img_eq = clahe_equalization(img, clip_limit=3.0)

# Visualização
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
ax1.imshow(img, cmap='gray')
ax1.set_title('Radiografia Original', fontsize=14)
ax1.axis('off')

ax2.imshow(img_eq, cmap='gray')
ax2.set_title('Após CLAHE (contraste realçado)', fontsize=14)
ax2.axis('off')

plt.tight_layout()
plt.show()
```
